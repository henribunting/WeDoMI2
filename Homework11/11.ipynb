(b) Implement a one-dimensional self-organizing map (Kohonen network, online algorithm for SOMs) using a Gaussian neighborhood function


$$hqp = e ^{−\frac{(q−p)^2 }{ 2σ^2}}$$


```python
def hqp(q, p, sigma):
    return np.exp(- np.power(q-p, 2)/np.power(sigma, 2))

def distance(a, b):
    return np.sqrt(np.sum(np.power(b-a, 2)))

def p(xa, w):
    # p = argmin q ( |Xa - woldq|)
    d = np.zeros((w.shape[0]))
    for q in range(w.shape[0]):
        d[q] = distance(xa, w[q,:])
    myp = np.argmin(d)
    
    return myp

def SOM(M, epsilon, sigma, x):
    w = np.random.rand(M, x.shape[1])
    for q in range(M):
        a = np.random.randint(x.shape[0])
        w[q,:] = w[q,:]+x[a,:]
    return SOMw(M, epsilon, sigma, x, w)
    
def SOMw(M, epsilon, sigma, x, w):
    #Initialization
    #M = number of partitions
    #Choose annealing schedule for epsilon and sigma
    eta = .99
    #Initialize Prototypes
    
    #
    maxiter = 10000
    
    #Optimization
    i = 0
    while i < maxiter:
        #choose a data point
        a = np.random.randint(x.shape[0])
        
        #determine closest prototype
        myp = p(x[a,:],w)
        
        #update prototypes
        dw = np.zeros((M, x.shape[1]))
        for q in range(M):
            dw[q,:] = epsilon*hqp(q, myp, sigma)*(x[a,:] - w[q,:])
        
        #print(np.sum(np.absolute(dw)))
        
        w = w + dw
        
        sigma = sigma*eta
        i = i + 1
    
    assignments = np.zeros(x.shape[0])
    for a in range(x.shape[0]):
        assignments[a]=p(x[a,:],w)
    
    return w, sigma, assignments
```

```python
SOM(4, .01, 1, x)
```

(d) Plot the final map in the data space, i.e. the locations of the prototypes and their connections, for each number of nodes k.

```python
for k in [4, 8, 16, 32, 64, 128]:
    plt.figure(figsize=(20,10))
    plt.xlim((0,2))
    plt.ylim((0,1))
    w, sigma, assignments = SOM(k, .01, 1, x)
    #print(sigma)
    #print(w)
    for K in range(k):
        xxK = xx[assignments==K]
        xyK = xy[assignments==K]
        plt.scatter(xxK, xyK, c=np.random.rand(3,1), alpha=1, marker="o", linewidth=0, s=70)
    plt.scatter(w[:,0], w[:,1], marker="x", color="magenta", s=100)
    plt.show()
```
